```python

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re

```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python

stop_words = set(stopwords.words('english'))
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    words = text.split()
    words = [word for word in words if word not in stop_words]
    words = [lemmatizer.lemmatize(word) for word in words]
    return " ".join(words)
```

```python

df['tags'] = df['tags'].apply(preprocess)
```

```python

df = df.reset_index(drop = True)
```

```python
indices = pd.Series(df.index,index = df['title']).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfi = TfidfVectorizer(stop_words='english',max_features=50000,ngram_range=(1,2))
```

```python

tfidf_matrix = tfi.fit_transform(df['tags'])
```

```python

tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n = 10):
    if title not in indices:
        return ['Movie not found']

    idx = indices[title]
    sim_scores = cosine_similarity(tfidf_matrix[idx],tfidf_matrix ).flatten()
    sim_idx = sim_scores.argsort()[::-1][1:n+1]
    return df['title'].iloc[sim_idx]
```

```python

recommend('',10)
```
